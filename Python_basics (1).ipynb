#Python basic questions


1. What is python and why it is popular?
   - Python is a high-level, interpreted, object-oriented programming language known for its simple and readable syntax.It is popular because
   Easy to read and learn,Large standard library,Supports multiple programming styles (OOP, procedural, functional),Strong community support,Widely used in data science, AI, web development, automation, etc.

2. What is an interpreter in Python?
   - An interpreter is a program that executes Python code line by line.It translates human-written Python code into machine-understandable instructions at runtime.   

3. What are pre-defined keywords in Python?
   - Keywords are reserved words that have special meaning in Python.Examples: if, else, for, while, class, True, False, etc.

4. Can keywords be used as variable names?
   - No.Keywords cannot be used as variable names because Python reserves them for its own syntax rules.

5. What is mutability in Python?
   - Mutability means the ability of an object to be changed after creation.

6. Why are lists mutable, but tuples are immutable?
   - Lists are designed for dynamic data; Python stores them in a way that allows element modification.
  Tuples are designed for fixed data; Python stores them in a way that prevents modifications, increasing efficiency and safety.

7. Difference between “==” and “is” operators in Python?
   - ==	Checks if values are equal.Example	[1] == [1] → True
  is	Checks if both references point to the same object in memory. Exmaple [1] is [1] → False

8. What are logical operators in Python?
   - Logical operators are used to combine conditional statements:
  and → True if both conditions are True
  or → True if at least one condition is True
  not → Reverses the condition

9. What is type casting in Python?
   - Type casting is the process of converting one data type into another, such as converting a string to an integer.

10. Difference between implicit and explicit type casting
    - Implicit casting	Python automatically converts the type.
   Example	3 + 5.0    →  8.0
   Explicit casting	Programmer manually converts types. Example	int("5")

11. Purpose of conditional statements in Python
    - Conditional statements allow Python to make decisions and execute different code depending on conditions.

12. How does the elif statement work?
    - elif stands for else-if.
  It lets you check multiple conditions, one after another.

13. Difference between for and while loops?
    - For Loop: Used when the number of iterations is known,Iterates over sequences,Example: looping over a list
   While Loop: Used when the number of iterations is unknown,Runs until a condition becomes False,Example: waiting for user input

14. Scenario where a while loop is more suitable than a for loop?

    - A while loop is better when you don’t know in advance how many times the loop should run.
  Example:Keep asking for a password until it's correct

10. Write code to demonstrate type casting with list elements?

```python
str_list = ["1", "2", "3", "4"]

int_list = list(map(int, str_list))
float_list = list(map(float, str_list))

print("Original list:", str_list)
print("As integers:", int_list)
print("As floats:", float_list)
```

13. Write a Python program to find the sum of all even numbers between 1 and 50?

```python
total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i

print("Sum of even numbers from 1 to 50 is:", total)
```

14. Write a program to reverse a string using a while loop?

```python
s = input("Enter a string: ")
reverse = ""
i = len(s) - 1

while i >= 0:
    reverse += s[i]
    i -= 1

print("Reversed string:", reverse)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop?

```python
num = int(input("Enter a number: "))
factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)
```
